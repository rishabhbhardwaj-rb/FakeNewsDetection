```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset_NER = pd.read_csv("/content/drive/MyDrive/FNDS/Dataset/FeatureExtraction/dataset_NER.zip", encoding='utf-8')
dataset_POSTag = pd.read_csv("/content/drive/MyDrive/FNDS/Dataset/FeatureExtraction/dataset_POSTag.zip", encoding='utf-8')
dataset_Dependency = pd.read_csv("/content/drive/MyDrive/FNDS/Dataset/FeatureExtraction/dataset_Dependency.zip", encoding='utf-8')
dataset_Sentiment = pd.read_csv("/content/drive/MyDrive/FNDS/Dataset/FeatureExtraction/dataset_Sentiment.zip", encoding='utf-8')
dataset_Cleaned = pd.read_csv("/content/drive/MyDrive/FNDS/Dataset/CleanedDataset/Dataset_Clean.zip",encoding='utf-8')
```

```python
combined_data =  pd.concat([dataset_Cleaned, dataset_Sentiment, dataset_Dependency, dataset_POSTag, dataset_NER],axis=1)
```

```python
combined_data.head()
```

```python
train_data, test_data = train_test_split(combined_data, test_size=0.3, random_state=40)
# train_data, valid_data = train_test_split(train_data, test_size=0.1, random_state=40)
```
